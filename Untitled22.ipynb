```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import seaborn as sns

def build_graph():
    G = nx.Graph()
    print("\nEnter edges in the format: source destination weight")
    print("Type 'done' when finished(keep in mind to enter 'done' in small words).\n")

    while True:
        line = input("Let's design your own city (source destination weight): ").strip()
        if line.lower() == 'done':
            break
        try:
            u, v, w = line.split()
            G.add_edge(u, v, weight=float(w))
        except ValueError:
            print("Invalid input. Please use the format: node1 node2 weight")
    return G

def find_best_route(G, source, destination):
    try:
        path = nx.dijkstra_path(G, source, destination, weight='weight')
        cost = nx.dijkstra_path_length(G, source, destination, weight='weight')
        return path, cost
    except nx.NetworkXNoPath:
        return [], None
    except nx.NodeNotFound as e:
        print(f"\n Node error: {e}")
        return None, None

def draw_graph(G, path=None):
    pos = nx.spring_layout(G, seed=42)
    edge_weights = nx.get_edge_attributes(G, 'weight')

    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=10)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights)
    if path:
        path_edges = list(zip(path, path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)
    plt.title("Graph with Shortest Route (in Red)")
    plt.show()

def generate_heatmap():
    data = [[random.randint(5, 20) for _ in range(5)] for _ in range(5)]
    sns.heatmap(data, annot=True, cmap="Reds", cbar=True)
    plt.title("Average Travelling Time (in minutes)")
    plt.show()

def main():
    print(" Let's find the best route to meet your friend - Custom Graph Route Finder")
    G = build_graph()
    print("\n Nodes in your map:", list(G.nodes))

    attempt = 0
    max_attempts = 2

    while attempt < max_attempts:
        source = input("Enter source node: ").strip()
        destination = input("Enter destination node: ").strip()

        path, cost = find_best_route(G, source, destination)

        if path is None:
            attempt += 1
            if attempt < max_attempts:
                print("\n Invalid node(s). Please re-enter valid source and destination.")
            else:
                print("\n Could not compute route due to invalid node(s).")
                draw_graph(G)
        elif not path:
            print(f"\n Sorry but you cannot reach your selected destination. Please select some other destination.\n No route exists between '{source}' and '{destination}'.")
            print("Please ensure there is a valid connection. Here's your current graph:")
            draw_graph(G)
            break
        else:
            print("\n Best Route:", " -> ".join(path))
            print("Total Travel Time:", cost)
            draw_graph(G, path)
            break

    generate_heatmap()

if __name__ == "__main__":
    main()
```
